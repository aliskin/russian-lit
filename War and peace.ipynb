```python
with open('voina-i-mir-1-1.txt') as f:
    text = f.readlines()
```

```python
2 < 5 < 10
```

```python
pieces = []
current_len = 0
LIMIT = 250
current_piece = []
for paragraph in text:
    doc = nlp(''.join(paragraph))
    if current_len + len(doc) <= LIMIT:
        current_len += len(doc)
        current_piece.append(paragraph)
    else:
        pieces.append(current_piece)
        current_piece = [paragraph]
        current_len = len(doc)
```

```python
for piece in pieces:
    print(piece)
    print()
```
