```python
import spacy
nlp = spacy.load("ru_core_news_sm")
```

```python
import pandas as pd
```

```python
with open('data/voina-i-mir-1-1.txt') as f:
    text = f.readlines()
```

```python
pieces = []
current_len = 0
LIMIT = 300
current_piece = []
for paragraph in text:
    paragraph = ' '.join(paragraph.replace('[', '').replace(']', '').split())
    doc = nlp(''.join(paragraph))
    if current_len + len(doc) <= LIMIT:
        current_len += len(doc)
        current_piece.append(paragraph)
    else:
        pieces.append(current_piece)
        current_piece = [paragraph]
        current_len = len(doc)
if current_piece:
    pieces.append(current_piece)
```

```python
for piece in pieces:
    for line in piece:
        print(line)
    print()
```

```python
len(pieces)
```

```python
for piece in pieces:
    doc = nlp(''.join(piece))
    print(len(doc))
    print(len(doc.ents), doc.ents)
```

```python
df = pd.DataFrame({'INPUT:text': ['\n'.join(piece) for piece in pieces]})
df.to_csv('data/voina-i-mir-1-1.tsv', sep='\t', index=False)
```

```python
def load_and_split(path, limit=300):
    with open(path) as f:
        text = f.readlines()
    pieces = []
    current_len = 0
    LIMIT = 300
    current_piece = []
    for paragraph in text:
        doc = nlp(''.join(paragraph))
        if current_len + len(doc) <= LIMIT and (not paragraph.startswith('Глава')):
            current_len += len(doc)
            current_piece.append(paragraph.strip())
        else:
            if current_piece:
                pieces.append(current_piece)
            current_piece = [paragraph.strip()]
            current_len = len(doc)
    if current_piece:
        pieces.append(current_piece)
    return pieces
```

```python
pieces = load_and_split('data/anna-karenina.txt')
```

```python
df = pd.DataFrame({'INPUT:text': ['\n'.join(piece) for piece in pieces for i in range(2)]})
df.to_csv('data/anna-karenina.tsv', sep='\t', index=False)
```
