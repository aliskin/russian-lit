```python
def load_and_split(path, limit=300):
    with open(path) as f:
        text = f.readlines()
    pieces = []
    current_len = 0
    LIMIT = 300
    current_piece = []
    for paragraph in text:
        if not paragraph.strip():
            continue
        doc = nlp(''.join(paragraph))
        if current_len + len(doc) <= limit and (not paragraph.startswith('Глава')):
            current_len += len(doc)
            current_piece.append(paragraph.strip())
        else:
            if current_piece:
                pieces.append(current_piece)
            current_piece = [paragraph.strip()]
            current_len = len(doc)
    if current_piece:
        pieces.append(current_piece)
    return pieces
```

```python
df_post = df.rename({
    'OUTPUT:result': 'INPUT:input_data',
    'ASSIGNMENT:assignment_id': 'INPUT:assignment_id'
}, axis=1).sample(n=20, random_state=31)
df_post.drop(['OUTPUT:no_per', 'GOLDEN:no_per', 'GOLDEN:result'], axis=1, inplace=True)
df_post.head(1)
```

```python
import json
```

```python
for _, row in df_post.iterrows():
    start = 0
    positions = []
    while '\r\n' in row['INPUT:text'][start:]:
        start = row['INPUT:text'].find('\r\n', start)
        positions.append(start)
        start += 1
    cur = 0
    labels = json.loads('[' + row['INPUT:input_data'].replace('\\', '') + ']')
    print(positions)
    if not positions:
        continue
    for label in labels:
        if cur < len(positions) and label['offset'] < positions[cur]:
            continue
        while cur < len(positions) and positions[cur] < label['offset']:
            cur += 1
        label['offset'] -= cur
        # print(label)
    print(labels)
    print(str(labels))
```
